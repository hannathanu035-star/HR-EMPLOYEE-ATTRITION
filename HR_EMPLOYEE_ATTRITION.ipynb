```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("/content/WA_Fn-UseC_-HR-Employee-Attrition (1).csv")

# Display first 5 rows
print(df.head())
```

```python
print(df.info())
```

```python
print(df.isnull().sum())
```

```python
print(df["Attrition"])
```

```python
print("Average age:",np.mean(df["Age"]))
```

```python
print("Average Monthly Income:",np.mean(df["MonthlyIncome"]))
```

```python
jobrole_income=df.groupby("JobRole")["MonthlyIncome"].mean()
print(jobrole_income)
```

```python
sns.countplot(x="Attrition",data=df)
plt.show()
```

```python
plt.hist(df["Age"],bins=10)


plt.title("Distribution of Age",)
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
sns.boxplot(x="Attrition",y="MonthlyIncome",data=df)
plt.title("Salary vs Attrition")
plt.show()
```

```python
sns.heatmap(df.corr(numeric_only=True),annot=False)
plt.title("correlation heatmap")
plt.show()
```
